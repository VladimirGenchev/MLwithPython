```python
def random_forest_features(model, X):
    importances = model.feature_importances_
    std = np.std([model.feature_importances_ for tree in model.estimators_],
                 axis=0)
    indices = np.argsort(importances)[::-1]

    plt.figure(figsize=(10, 10))
    plt.title("Feature importances")
    plt.bar(range(X.shape[1]), importances[indices], color="r", yerr=std[indices], align="center")
    plt.xticks(range(X.shape[1]), X.columns, rotation='vertical')
    plt.xlim([-1, X.shape[1]])
    plt.show()
```

```python
random_forest_features(model, X)
```

Може би тестовите данни, които се опитваме да предскажем съдържат информация за хора, които са с по-концентрирана средна възраст над тази от тренировъчните данни (ПРИМЕРНО!). Или пък може данните да са наситени с повече хора, които например са мъже и затова за алгоритъма статистически да е по-трудно да уцели повече верни попадения. Нека да разлгедаме съотношението на тези параметри между сета, с който сме тренирали и между тестовия сет от другия файл.

```python
print("{}% from training set are under 20 years old, and {}% from the people from test set are under 20 years old.".format(X_train[X_train.Age < 20].shape[0] / X_train.shape[0], test[test.Age < 20].shape[0] / test.shape[0]))
```

```python
print("{}% from training set are over 50 years old, and {}% from the people from test set are over 50 years old.".format(X_train[X_train.Age > 50].shape[0] / X_train.shape[0], test[test.Age > 50].shape[0] / test.shape[0]))
```

```python
print("{}% from training set are males, and {}% from the people from test set are males.".format(X_train[X_train.Male == 1.0].shape[0] / X_train.shape[0], test[test.Male == 1.0].shape[0] / test.shape[0]))
```

```python
print("{}% from training set are 1st class, and {}% from the people from test set are 1st class.".format(X_train[X_train.Pclass == 1].shape[0] / X_train.shape[0], test[test.Pclass == 1].shape[0] / test.shape[0]))
```

Единствената малка разлика е че хората, които са на възраст под 20 години в тренировъчния сет са малко повече, отколкото в тестовия. Това теоретично може да се отрази на резултатите, но като цяло изглежда, че съотношението при повечето показатели е запазено.

Никъде в горните примери не опитахме да използваме DecisionTreeClassifier. Нека да опитаме с него:

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier().fit(X_train, y_train)
print("train score:", classifier.score(X_train, y_train))
print("test score: ", classifier.score(X_test, y_test))
```

Тук резултатът при тренировъчния сет е доста добър, но пък алгоритъмът силно овърфитва. Може да опитаме да си поиграем с параметъра min_samples_split, който служи за регуляризация в този случай.

```python
for min_samples in [2, 3, 5, 8, 10, 11, 12, 13, 15, 17, 20, 25]:
    classifier = DecisionTreeClassifier(min_samples_split = min_samples).fit(X_train, y_train)
    
    print("Min_samples = {:f} train: {:f} test: {:f}".format(min_samples, classifier.score(X_train, y_train), classifier.score(X_test, y_test)))
```

Виждаме, че при min_samples_split = 11 се постига доста добър резултат - почти 84% успеваемост за тестовите данни. В този случай, въпреки че все още не е много добре, вече успешно получихме по-добри резултати от по-горе получените и с този опит вече задачата е изпълнена успешно.
